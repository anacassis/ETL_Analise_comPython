```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud 
```

```python
frase = 'As pessoas inventam estatísticas para provar qualquer coisa. 40% das pessoas sabem disso! A estatística pode ser definida como um conjunto de métodos utilizados para tomada de decisões sábias em face à incerteza. Estatística: a ciência que diz que se eu comi um frango e tu não comeste nenhum, teremos comido, em média, meio frango cada um.'
frase = frase + ' Os erros causados por dados inadequados são muito menores do que aqueles devido à falta total de dados.Uma análise estatística, conduzida adequadamente, é uma dissecação delicada de incertezas, uma cirurgia de suposições.'
frase
```

```python
# Gerar nuvem de palavras

nuvem_palavras = WordCloud(
    background_color='black',
    width=1900,
    height=800).generate(frase)
```

```python
# Criar o gráfico 

fig, ax = plt.subplots( figsize=(10,7))

ax.imshow( nuvem_palavras, interpolation = 'bilinear')

ax.set_axis_off()  # retira eixos internos 
```
