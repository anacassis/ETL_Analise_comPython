```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.graph_objects as Dash

```

```python
dados = pd.read_excel('Dados_Bitcoin.xlsx')
dados
```

```python
# Setar o index
dados.set_index("Date", inplace=True)
```

```python
# GrÃ¡fico de linhas 
fig = px.line(dados, y = 'Close')

# plotar 
fig.show()
```

```python
media_movel = dados['Close'].rolling(5).mean()
media_tendencia = dados['Close'].rolling(30).mean()
```

```python
# Criar dash 
figura = Dash.Figure()

# Primeiro Eixo
figura.add_trace(
    Dash.Scatter(
        x= dados.index,
        y=dados.Close,
        mode ='lines',
        name='Fechamento',
        marker_color='#F08080'
    )
)

# Segundo Eixo
figura.add_trace(
    Dash.Scatter(
        x= dados.index,
        y=media_movel,
        mode ='lines',
        name='MÃ©dia Movel',
        marker_color='#9370DB'
    )
)

# Terceiro Eixo
figura.add_trace(
    Dash.Scatter(
        x= dados.index,
        y=media_tendencia,
        mode ='lines',
        name='TendÃªncia',
        marker_color='#ADFF2F'
    )
)

# Ajustando o layout
figura.update_layout(
    
    title='AnÃ¡lise do Fechamento do Bitcoin',
    titlefont_size=20,

    xaxis = dict(
        title='PerÃ­odo',
        titlefont_size=14,
        tickfont_size=10),
    
    yaxis = dict(
        title='PreÃ§o Fechamento ($)',
        titlefont_size=14,
        tickfont_size=10
    ),
    # Parametros para Legenda 
    legend = dict(
        y=-0.2,
        x=-0.1  ,
        
    )
    )
```
