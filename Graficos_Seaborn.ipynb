```python
import numpy as np
import pandas as pd
import seaborn as sns

# Definindo tema(não é necessario)
sns.set_theme(style = 'darkgrid')
```

```python
#Dados
base_dados = sns.load_dataset('tips')

# Verificar dados
base_dados.head()
```

```python
# Renomeando colunas 

base_dados.rename(columns = {'total_bill':'total_conta',
                             'tip':'gorjeta',
                             'sex':'sexo',
                             'smoker':'fumante',
                             'day':'dia',
                             'time':'periodo',
                             'size':'pessoas_namesa'}, inplace=True)
display(base_dados)
```

```python
# Gráfico Relplot
sns.relplot(x='total_conta',y='gorjeta', color='green', data=base_dados);
```

```python
# Gráfico relplot
# Passando outro parâmetro como classe no gráfico
sns.relplot(x='total_conta', y='gorjeta', data = base_dados, hue = 'sexo');
```

```python
sns.relplot(x='total_conta', y='gorjeta', data = base_dados, kind='line', color='orange');
```

```python
sns.relplot(x='total_conta', y='gorjeta', data = base_dados, kind='line', hue = 'fumante');
```

```python
# Histograma
sns.histplot(data = base_dados, x='total_conta', color ='green');
```

```python
# Histograma com classes
sns.histplot(data = base_dados, x='total_conta', hue= 'fumante');
```

```python
# Barplot grafico de barras com classe
sns.barplot(data = base_dados, x = 'sexo', y='total_conta', hue = 'fumante', color = 'orange');
```

```python
sns.pairplot(data= base_dados);
```

```python
sns.pairplot(data = base_dados, hue = 'sexo');
```

```python
sns.pairplot(data = base_dados, kind = 'kde');
```

```python
sns.boxplot( data = base_dados, x = 'dia', y ='total_conta');
```

```python
sns.boxplot( data = base_dados, x = 'dia', y ='total_conta', hue = 'sexo');
```
