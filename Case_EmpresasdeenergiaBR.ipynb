```python
import seaborn as seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados = pd.read_excel('Dados_Empresas_Energia.xlsx')
dados
```

```python
dados.set_index('Data', inplace=True)
```

```python
dados.head()
```

```python
# Criar o gráfico 
# Estilo
plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(16,7))
plt.title('Análise ações de empresas de Energia', loc='left', fontsize=12, fontweight=0)

plt.plot(dados.index,dados['Petrobras'], color = '#008000', linewidth=3, alpha=0.7)# Alpha clarear a cor da linha

# Texto Petrobras
plt.text(dados.index[-1], dados['Petrobras'].tail(1),'Petrobras', color='#008000', size='large', horizontalalignment='left')

for Colunas in dados.columns[1:]:
  plt.plot(dados.index,dados[Colunas], color = 'gray', linewidth=3, alpha=0.7)# Alpha clarear a cor da linha
  plt.text(dados.index[-1], dados[Colunas].tail(1),Colunas, color='gray', size='large', horizontalalignment='left')


plt.xlabel('Período')
plt.ylabel('Preço de Fechaamento (R$)')
```
