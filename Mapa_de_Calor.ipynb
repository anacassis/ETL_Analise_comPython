```python
import pandas as pd 
import plotly.express as px
```

```python
#
cidade = ['Santo André', 'São Bernardo do Campo','Diadema', 'São Caetano do Sul',
           'Maua','Ribeirão pires','Rio Grande da Serra','São Paulo']

estado = ['São Paulo','São Paulo','São Paulo','São Paulo','São Paulo',
          'São Paulo','São Paulo','São Paulo',]

# Gerando a Latitude
Latitude = [-23.6666, -23.6944, -23.6865, -23.6226, -23.6687, 
            -23.7141, -23.7452, -23.5489]

# Gerando a Longitude
Longitude = [-46.5322, -46.5654, -46.6234, -46.5489, -46.4614, 
             -46.4137, -46.4022, -46.6388]

# Valor
Vendas = [100, 120, 90, 50, 70, 90, 250, 400]
```

```python
# Dicionário 
dic = {
    'cidade' : cidade,
    'estado' : estado,
    'latitude' : Latitude,
    'longitude' : Longitude,
    'vendas' : Vendas
}
base_dados = pd.DataFrame(dic)
base_dados
```

```python
# Mapa de Calor Geográfica 

px.density_mapbox(
    base_dados,
    lat='latitude',
    lon='longitude',
    z='vendas',
    radius=30,
    center=dict(lat=-2.3700, lon=-46.5555),
    zoom=8,
    mapbox_style = 'stamen-terrain'
)
```
